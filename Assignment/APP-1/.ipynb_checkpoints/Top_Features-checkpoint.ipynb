```python
# IP-based features
def ip_to_int(ip):
    try:
        parts = [int(part) for part in ip.split('.')]
        return parts[0] * 256**3 + parts[1] * 256**2 + parts[2] * 256 + parts[3]
    except:
        return 0
```

```python
def transform_ip(df):
    # df['Source_IP_num'] = df['Source_IP'].apply(ip_to_int)
    df['Destination_IP_num'] = df['Destination_IP'].apply(ip_to_int)
```

```python
# Identify columns with string data
string_columns = X_train_drop.select_dtypes(include=['object', 'string']).columns

print(string_columns)

# Initialize a dictionary to store LabelEncoders
label_encoders = {}

# Apply LabelEncoder to each string column in both X_train and X_test
for column in string_columns:
    
    # Fit on X_train and transform both X_train and X_test
    X_train_drop[column] = label_encoder.fit_transform(X_train_drop[column])
    X_test_drop[column] = label_encoder.transform(X_test_drop[column])
    
    # Store the encoder for future reference
    label_encoders[column] = label_encoder

# # Print the transformed DataFrames
# print("Transformed X_train:")
# print(X_train_drop)
# print("\nTransformed X_test:")
# print(X_test_drop)
```

```python
# # Verify all data is numerical
# assert X_train.select_dtypes(include=['object', 'string']).empty, "X_train still has non-numeric data!"
# assert X_test.select_dtypes(include=['object', 'string']).empty, "X_test still has non-numeric data!"
```

```python
# Find columns with non-numeric data in X_test
# non_numeric_columns = X_test_drop.select_dtypes(include=['object', 'string']).columns
non_numeric_columns = X_test_drop.select_dtypes(include=['object', 'string']).columns

# Print the non-numeric columns
print("Non-numeric columns in X_test:")
print(non_numeric_columns)
```

```python
for column in non_numeric_columns:
    if column in label_encoders:  # Check if this column was encoded in X_train
        le = label_encoders[column]
        X_test_drop[column] = le.transform(X_test_drop[column])
    else:
        print(f"Warning: {column} was not encoded in X_train!")
```

```python
# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_drop)
X_test_scaled = scaler.transform(X_test_drop)
```

```python
# BaggingClassifier with DecisionTree as base estimator
bagging_clf = BaggingClassifier(random_state=28, n_estimators=150,max_samples=0.75, max_features=0.55)
bagging_clf.fit(X_train_scaled, y_train_encoded)
y_pred_bagging = bagging_clf.predict(X_test_scaled)
accuracy_bagging = accuracy_score(y_test_encoded, y_pred_bagging)
print(f"BaggingClassifier Accuracy on Test Dataset: {accuracy_bagging:.4f}")
print("Classification Report (BaggingClassifier):\n", classification_report(y_test_encoded, y_pred_bagging))
```

```python
rf = RandomForestClassifier(random_state=267, n_jobs=-1,n_estimators = 100)
rf.fit(X_train_scaled, y_train_encoded)
y_pred_rf = rf.predict(X_test_scaled)
accuracy_rf = accuracy_score(y_test_encoded, y_pred_rf)
print(f"RadomForest Accuracy on Test Dataset: {accuracy_rf:.4f}")
print("Classification Report (RadomForest):\n", classification_report(y_test_encoded, y_pred_rf))
```

```python
# Find pairs with a match of 3 or more
matches = []
for actual in cm_df.index:
    for predicted in cm_df.columns:
        if cm_df.loc[actual, predicted] >= 3:  # Check for matches >= 3
            if actual != predicted:
                matches.append((actual, predicted, cm_df.loc[actual, predicted]))
```

```python

# Print the matches
print("Pairs with matches of 3 or more:")
for actual, predicted, count in matches:
    print(f"Actual: {actual}, Predicted: {predicted}, Count: {count}")
```

```python
per_class_accuracy = cm.diagonal() / cm.sum(axis=1)
print("Per-Class Accuracy:")
print(per_class_accuracy)

overall_accuracy = np.trace(cm) / cm.sum()
print(f"Overall Accuracy: {overall_accuracy:.2%}")
```

```python
# Get the misclassified pairs
misclassified_pairs = []
for actual, predicted, count in matches:
    misclassified_pairs.append((actual, predicted))

# Create a new dataset with only the rows corresponding to the misclassified classes
filtered_train = pd.DataFrame()
filtered_test = pd.DataFrame()
for actual, predicted in misclassified_pairs:
    mask = train_df['label'] == actual
    filtered_rows = train_df[mask]
    filtered_train = pd.concat([filtered_train, filtered_rows])

# Filter test dataset for misclassified pairs
filtered_test = pd.DataFrame()
for actual, predicted in misclassified_pairs:
    mask = test_df['label'] == actual
    filtered_rows = test_df[mask]
    filtered_test = pd.concat([filtered_test, filtered_rows])

# Reset index of the new dataset
filtered_train = filtered_train.reset_index(drop=True)

print(f"Original training set size: {len(train_df)}")
print(f"Filtered training set size: {len(filtered_train)}")
print("\nClass distribution in filtered dataset:")
print(filtered_train['label'].value_counts())

# Save the filtered dataset if needed
filtered_train.to_csv('filtered_train.csv', index=False)


filtered_test = filtered_test.reset_index(drop=True)

print(f"Original test set size: {len(test_df)}")
print(f"Filtered test set size: {len(filtered_test)}")
print("\nClass distribution in filtered test dataset:")
print(filtered_test['label'].value_counts())

filtered_test.to_csv('filtered_test.csv', index=False)
```

```python
# Prepare filtered datasets
X_filtered_train = filtered_train[features]
X_filtered_test = filtered_test[features]

# Transform IPs
transform_ip(X_filtered_train)
transform_ip(X_filtered_test)

# Drop IP columns
X_filtered_train_drop = X_filtered_train.drop(columns=['Destination_IP'])
X_filtered_test_drop = X_filtered_test.drop(columns=['Destination_IP'])

# Scale features
X_filtered_train_scaled = scaler.transform(X_filtered_train_drop)
X_filtered_test_scaled = scaler.transform(X_filtered_test_drop)

# Encode labels
y_filtered_train = label_encoder.transform(filtered_train['label'])
y_filtered_test = label_encoder.transform(filtered_test['label'])

# Train new model
new_rf = RandomForestClassifier(random_state=42, n_jobs=-1, n_estimators=150)
new_rf.fit(X_filtered_train_scaled, y_filtered_train)
y_pred_new = new_rf.predict(X_filtered_test_scaled)

# Print comparison
print("Original Model Performance on Full Test Set:")
print(classification_report(y_test_encoded, y_pred_rf))

print("\nNew Model Performance on Filtered Test Set:")
print(classification_report(y_filtered_test, y_pred_new))

# Compare specific classes
misclassified_classes = set([x[0] for x in misclassified_pairs])
print("\nDetailed comparison for previously misclassified classes:")
for class_name in misclassified_classes:
    class_idx = label_encoder.transform([class_name])[0]
    
    # Original model metrics
    orig_mask = y_test_encoded == class_idx
    if orig_mask.any():
        orig_accuracy = accuracy_score(y_test_encoded[orig_mask], y_pred_rf[orig_mask])
        print(f"\nClass {class_name}:")
        print(f"Original model accuracy: {orig_accuracy:.4f}")
    
    # New model metrics
    new_mask = y_filtered_test == class_idx
    if new_mask.any():
        new_accuracy = accuracy_score(y_filtered_test[new_mask], y_pred_new[new_mask])
        print(f"New model accuracy: {new_accuracy:.4f}")
```
