```python
first_selection_threshold = 0.95
second_selection_lower_threshold = 0.85
second_selection_upper_threshold = 0.97
## First selection: Features with same value in 97% or more of the data
def find_constant_features(df, threshold=0.97):
    constant_features = []
    for column in df.columns:
        # Calculate the proportion of the most frequent value
        most_frequent_prop = df[column].value_counts(normalize=True).iloc[0]
        if most_frequent_prop >= threshold:
            constant_features.append(column)
    
    return constant_features

# Second selection: Check distribution of features with same value in 85-97% of data
def analyze_near_constant_features(df, y, lower_threshold=0.85, upper_threshold=0.98):
    features_to_remove = []
    total_classes = len(np.unique(y))
    
    print(f"Total number of classes in dataset: {total_classes}")
    
    # Expected uniform distribution probability
    uniform_prob = 1.0 / total_classes
    # Allow some deviation from uniform (e.g., 2x the uniform probability)
    max_allowed_prob = uniform_prob * 2
    
    for column in df.columns:
        # Get value counts and proportions
        value_counts = df[column].value_counts(normalize=True)
        most_frequent_prop = value_counts.iloc[0]
        most_frequent_value = value_counts.index[0]
        
        # If feature has same value in 85-97% of data
        if lower_threshold <= most_frequent_prop < upper_threshold:
            # Get the minority cases (where value is different from most frequent)
            minority_mask = df[column] != most_frequent_value
            minority_labels = y[minority_mask]
            
            # Check distribution of classes in minority cases
            class_distribution = minority_labels.value_counts(normalize=True)
            
            # Analyze distribution
            if len(class_distribution) > 1:  # Only if there are multiple classes
                max_class_prop = class_distribution.max()
                
                # Calculate entropy of distribution
                entropy = -sum(p * np.log2(p) for p in class_distribution)
                max_entropy = -np.log2(1/total_classes)  # Maximum possible entropy
                entropy_ratio = entropy / max_entropy
                
                print(f"Entropy ratio: {entropy_ratio:.2f} (1.0 would be perfectly uniform)")
                
                # Decision criteria:
                # 1. If any class is heavily overrepresented
                # 2. Or if distribution is close to uniform
                if max_class_prop < max_allowed_prob or entropy_ratio > 0.8:
                    features_to_remove.append(column)
                    print("REMOVED: Distribution is too uniform or lacks clear class separation")
                else:
                    print("KEPT: Shows significant class separation")
            else:
                print("KEPT: Only one class in minority cases (strong predictor)")
                
    return features_to_remove

# Apply the feature selection
print("Starting feature selection...")
print(f"Number of unique classes in dataset: {len(np.unique(y))}")

# First stage: Remove highly constant features
constant_features = find_constant_features(X, first_selection_threshold)
print(f"\nFound {len(constant_features)} features with {first_selection_threshold}% constant values:")
for feat in constant_features:
    value_counts = X[feat].value_counts(normalize=True)
    print(f"{feat}: Most common value appears in {value_counts.iloc[0]:.2%} of cases")

# Remove these features
X = X.drop(columns=constant_features)
test_df = test_df.drop(columns=constant_features)
val_df = val_df.drop(columns=constant_features)

# Second stage: Analyze near-constant features
features_to_remove = analyze_near_constant_features(X, y, second_selection_lower_threshold, second_selection_upper_threshold)
print(f"\nFound {len(features_to_remove)} near-constant features to remove:")
for feat in features_to_remove:
    print(f"\n{feat}:")
    value_counts = X[feat].value_counts(normalize=True)
    print(f"Value distribution: {value_counts.head()}")

# Remove these features
X = X.drop(columns=features_to_remove)
test_df = test_df.drop(columns=features_to_remove)
val_df = val_df.drop(columns=features_to_remove)

print("\nFeature selection completed.")
print(f"Number of features after selection: {X.shape[1]}")
print("\nRemoved features summary:")
print(f"- {len(constant_features)} highly constant features {first_selection_threshold}")
print(f"- {len(features_to_remove)} near-constant features with uniform class distribution")
```

```python
def plot_confusion_matrix(y_true, y_pred, figsize=(20, 20), threshold=1, samples_per_class=5):
    """
    Create and plot a confusion matrix, and analyze chronic misclassifications.
    
    Parameters:
    -----------
    y_true : array-like
        True labels
    y_pred : array-like
        Predicted labels
    figsize : tuple
        Figure size for the plot (width, height)
    threshold : int
        Minimum number of misclassifications to consider it chronic
    samples_per_class : int
        Number of samples per class (default=5)
    """
    # Get unique labels (sorted to ensure consistent order)
    labels = sorted(list(set(list(y_true) + list(y_pred))))
    
    # Create the confusion matrix
    cm = confusion_matrix(y_true, y_pred, labels=labels)
    
    # Create confusion matrix DataFrame for analysis
    cm_df = pd.DataFrame(cm, index=labels, columns=labels)
    
    # Create figure and axes
    plt.figure(figsize=figsize)
    
    # Create heatmap
    sns.heatmap(
        cm,
        xticklabels=labels,
        yticklabels=labels,
        annot=True,
        fmt='g',
        cmap='Blues',
        square=True,
        cbar=True,
        annot_kws={'size': 8}
    )
    
    # Rotate x-axis labels
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    
    # Set labels and title
    plt.xlabel('Predicted Class')
    plt.ylabel('Actual Class')
    plt.title('Confusion Matrix')
    
    # Adjust layout to prevent label cutoff
    plt.tight_layout()
    
    # Show the plot
    plt.show()
    
    # Analyze chronic misclassifications
    print(f"\nAnalyzing chronic misclassifications (threshold >= {threshold}):\n")
    print("Format: ACTUAL_WEBSITE was misclassified as PREDICTED_WEBSITE X times\n")
    
    # Find and display chronic misclassifications
    found_misclassifications = False
    chronic_misclassifications = []
    
    for actual_site in labels:
        for predicted_site in labels:
            if actual_site != predicted_site:
                misclassification_count = cm_df.loc[actual_site, predicted_site]
                if misclassification_count >= threshold:
                    found_misclassifications = True
                    msg = f"MISS: '{actual_site}' was misclassified as '{predicted_site}' : {misclassification_count} times"
                    print(msg)
                    chronic_misclassifications.append({
                        'actual': actual_site,
                        'predicted': predicted_site,
                        'count': misclassification_count,
                        'percentage': (misclassification_count / samples_per_class) * 100
                    })
    
    if not found_misclassifications:
        print("No chronic misclassifications found with the current threshold.")
    
    # Display percentages for all misclassifications
    print("\nMisclassification Percentages (all cases):")
    print("-" * 50)
    for actual_site in labels:
        for predicted_site in labels:
            if actual_site != predicted_site:
                misclassification_count = cm_df.loc[actual_site, predicted_site]
                if misclassification_count > 0:
                    percentage = (misclassification_count / samples_per_class) * 100
                    print(f"{actual_site} â†’ {predicted_site}: {percentage:.1f}% ({misclassification_count}/{samples_per_class} samples)")
    
    # Find perfect classifications
    print("\nPerfect Classifications:")
    print("-" * 50)
    perfect_classes = []
    
    for label in labels:
        # Get row and column for this class
        row = cm_df.loc[label]  # All predictions for actual class
        col = cm_df[label]      # All predictions as this class
        
        # Check if classification is perfect:
        # 1. All actual samples were correctly classified (row sum equals samples_per_class and all in diagonal)
        # 2. No other samples were classified as this class (no false positives in column)
        if (row.sum() == samples_per_class and  # All samples accounted for
            row[label] == samples_per_class and  # All samples correctly classified
            col.sum() == samples_per_class):     # No false positives
            perfect_classes.append(label)
    
    if perfect_classes:
        print("\nThe following classes had perfect classification (no false positives or false negatives):")
        for cls in sorted(perfect_classes):
            print(f"- {cls}")
        print("\n Number of classes achieved perfect classification: ", len(perfect_classes))
    else:
        print("\nNo classes achieved perfect classification.")

    return chronic_misclassifications, perfect_classes
```

```python
def evaluate_improvement(y_true, y_pred_old, y_pred_new):
    """Evaluate if new predictions are actually better considering class balance"""
    old_accuracy = accuracy_score(y_true, y_pred_old)
    new_accuracy = accuracy_score(y_true, y_pred_new)
    
    # Get per-class performance
    old_report = classification_report(y_true, y_pred_old, output_dict=True)
    new_report = classification_report(y_true, y_pred_new, output_dict=True)
    
    # Calculate average per-class recall (to ensure balanced performance)
    old_recall_avg = np.mean([old_report[cls]['recall'] for cls in set(y_true)])
    new_recall_avg = np.mean([new_report[cls]['recall'] for cls in set(y_true)])
    
    # Only consider it an improvement if BOTH overall accuracy and per-class recall improved
    return (new_accuracy > old_accuracy and new_recall_avg >= old_recall_avg), {
        'accuracy_change': new_accuracy - old_accuracy,
        'recall_change': new_recall_avg - old_recall_avg
    }

def clean_and_scale_data(X_train, X_test, X_val=None):
    """Clean and scale data, handling infinities and large values"""
    # Create copies to avoid modifying originals
    X_train_clean = X_train.copy()
    X_test_clean = X_test.copy()
    X_val_clean = X_val.copy() if X_val is not None else None
    
    # Replace infinities with large but finite values
    X_train_clean = X_train_clean.replace([np.inf, -np.inf], np.nan)
    X_test_clean = X_test_clean.replace([np.inf, -np.inf], np.nan)
    if X_val_clean is not None:
        X_val_clean = X_val_clean.replace([np.inf, -np.inf], np.nan)
    
    # Fill NaN values with median of column
    for col in X_train_clean.columns:
        median = X_train_clean[col].median()
        X_train_clean[col] = X_train_clean[col].fillna(median)
        X_test_clean[col] = X_test_clean[col].fillna(median)
        if X_val_clean is not None:
            X_val_clean[col] = X_val_clean[col].fillna(median)
    
    # Scale the data
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train_clean)
    X_test_scaled = scaler.transform(X_test_clean)
    X_val_scaled = scaler.transform(X_val_clean) if X_val_clean is not None else None
    
    # Convert back to DataFrame to keep column names
    X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train_clean.columns)
    X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test_clean.columns)
    if X_val_scaled is not None:
        X_val_scaled = pd.DataFrame(X_val_scaled, columns=X_val_clean.columns)
        return X_train_scaled, X_test_scaled, X_val_scaled
    
    return X_train_scaled, X_test_scaled

def get_misclassified_groups(chronic_misclassifications):
    """
    Extract groups of related misclassifications.
    E.g., if facebook -> instagram and facebook -> twitter, 
    return [facebook, instagram, twitter] as one group
    """
    # Create a dictionary to store related classes
    class_groups = {}
    
    # Build groups of related classes
    for misclass in chronic_misclassifications:
        actual = misclass['actual']
        predicted = misclass['predicted']
        
        # If either class is already in a group, add both to that group
        found_group = False
        for group_id, group in list(class_groups.items()):
            if actual in group or predicted in group:
                group.add(actual)
                group.add(predicted)
                found_group = True
                # Check if this group should be merged with any other group
                for other_id, other_group in list(class_groups.items()):
                    if other_id != group_id and (actual in other_group or predicted in other_group):
                        group.update(other_group)
                        del class_groups[other_id]
                break
        
        # If neither class is in any group, create a new group
        if not found_group:
            class_groups[len(class_groups)] = {actual, predicted}
    
    # Convert to list of lists
    return [list(group) for group in class_groups.values()]

def knn_train_and_predict(X_train, y_train, X_test, k=5):
    """Train and predict using k-Nearest Neighbors"""
    # Initialize and train k-Nearest Neighbors model
    knn_model = KNeighborsClassifier(n_neighbors=k, n_jobs=-1)
    knn_model.fit(X_train, y_train)

    # Make predictions
    knn_predictions = knn_model.predict(X_test)
    
    return knn_predictions

def train_on_class_group(X, y, test_df, test_predicted_labels, val_df, val_predicted_labels, class_group,
                        social_media_features_train, social_media_features_test, social_media_features_val,
                        targeted_features_train, targeted_features_test, targeted_features_val,
                        enhanced_features_train, enhanced_features_test, enhanced_features_val):
    """Train and evaluate models on a group of related problematic classes"""
    
    # Filter data for this group of classes
    group_mask_train = y.isin(class_group)
    group_mask_test = test_predicted_labels.isin(class_group)
    group_mask_val = val_predicted_labels.isin(class_group)
    
    # Get all training data with reset indices
    X_group = X[group_mask_train].copy().reset_index(drop=True)
    y_group = y[group_mask_train].copy().reset_index(drop=True)
    test_df_group = test_df[group_mask_test].copy().reset_index(drop=True)
    test_labels_group = test_predicted_labels[group_mask_test].copy().reset_index(drop=True)
    val_df_group = val_df[group_mask_val].copy().reset_index(drop=True)
    
    # Combine all features into one DataFrame
    combined_train = X_group.copy()
    combined_test = test_df_group.copy()
    combined_val = val_df_group.copy()
    
    # Add social media features
    social_media_train = social_media_features_train[group_mask_train].copy().reset_index(drop=True)
    social_media_test = social_media_features_test[group_mask_test].copy().reset_index(drop=True)
    social_media_val = social_media_features_val[group_mask_val].copy().reset_index(drop=True)
    for col in social_media_train.columns:
        combined_train[f'social_{col}'] = social_media_train[col]
        combined_test[f'social_{col}'] = social_media_test[col]
        combined_val[f'social_{col}'] = social_media_val[col]
        
    # Add targeted features
    targeted_train = targeted_features_train[group_mask_train].copy().reset_index(drop=True)
    targeted_test = targeted_features_test[group_mask_test].copy().reset_index(drop=True)
    targeted_val = targeted_features_val[group_mask_val].copy().reset_index(drop=True)
    for col in targeted_train.columns:
        combined_train[f'targeted_{col}'] = targeted_train[col]
        combined_test[f'targeted_{col}'] = targeted_test[col]
        combined_val[f'targeted_{col}'] = targeted_val[col]
        
    # Add enhanced features
    enhanced_train = enhanced_features_train[group_mask_train].copy().reset_index(drop=True)
    enhanced_test = enhanced_features_test[group_mask_test].copy().reset_index(drop=True)
    enhanced_val = enhanced_features_val[group_mask_val].copy().reset_index(drop=True)
    for col in enhanced_train.columns:
        combined_train[f'enhanced_{col}'] = enhanced_train[col]
        combined_test[f'enhanced_{col}'] = enhanced_test[col]
        combined_val[f'enhanced_{col}'] = enhanced_val[col]
    
    # Clean and scale all data
    combined_train_clean, combined_test_clean, combined_val_clean = clean_and_scale_data(
        combined_train, combined_test, combined_val)
    
    val_predictions = None

    # handling groups containing: 'facebook', 'instagram' with KNN:
    if any(site in class_group for site in ['facebook', 'instagram']):
        # preparing the data set with only specific features:
        X_train_knn = combined_train_clean.copy()
        X_test_knn = combined_test_clean.copy()
        X_val_knn = combined_val_clean.copy()
        for col in combined_train_clean.columns:
            if 'pps_bwd' not in col and 'pps_fwd' not in col:
                X_train_knn = X_train_knn.drop(columns=[col])
                X_test_knn = X_test_knn.drop(columns=[col])
                X_val_knn = X_val_knn.drop(columns=[col])

        # Train and predict using k-Nearest Neighbors
        knn_predictions = knn_train_and_predict(X_train_knn, y_group, X_test_knn)
        val_predictions = knn_train_and_predict(X_train_knn, y_group, X_val_knn)
        knn_accuracy = accuracy_score(test_labels_group, knn_predictions)
        print(f"\nKNN Neighbors Accuracy: {knn_accuracy:.4f}")

        return test_labels_group, knn_predictions, knn_accuracy, confusion_matrix(test_labels_group, knn_predictions), val_predictions
    
    # handling groups containing: 'ca.gov' and 'amazonaws' with KNN:
    if any(site in class_group for site in ['ca.gov', 'amazonaws']):
        # preparing the data set with only specific features:
        X_train_knn = combined_train_clean.copy()
        X_test_knn = combined_test_clean.copy()
        X_val_knn = combined_val_clean.copy()
        for col in combined_train_clean.columns:
            if 'targeted_ca_amazon' not in col and 'max_bwd_packet' not in col:
                X_train_knn = X_train_knn.drop(columns=[col])
                X_test_knn = X_test_knn.drop(columns=[col])
                X_val_knn = X_val_knn.drop(columns=[col])

        # Train and predict using k-Nearest Neighbors
        knn_predictions = knn_train_and_predict(X_train_knn, y_group, X_test_knn)
        val_predictions = knn_train_and_predict(X_train_knn, y_group, X_val_knn)
        knn_accuracy = accuracy_score(test_labels_group, knn_predictions)
        print(f"\nKNN Neighbors Accuracy: {knn_accuracy:.4f}")

        return test_labels_group, knn_predictions, knn_accuracy, confusion_matrix(test_labels_group, knn_predictions), val_predictions

    # Train initial Random Forest to get feature importance
    rf_model = RandomForestClassifier(
        n_estimators=300,
        max_depth=None,
        min_samples_split=5,
        min_samples_leaf=2,
        max_features='sqrt',
        class_weight='balanced',
        random_state=238,
        n_jobs=-1
    )
    rf_model.fit(combined_train_clean, y_group)
    
    # Get feature importance
    feature_importance = pd.DataFrame({
        'feature': combined_train_clean.columns,
        'importance': rf_model.feature_importances_
    }).sort_values('importance', ascending=False)

    print("\nFeature importance for this group:")
    print(feature_importance.head(20))

    # predict using Random Forest:
    rf_predictions = rf_model.predict(combined_test_clean)
    val_predictions = rf_model.predict(combined_val_clean)
    best_predictions = rf_predictions
    best_val_predictions = val_predictions
    rf_accuracy = accuracy_score(test_labels_group, rf_predictions)
    best_accuracy = rf_accuracy
    print(f"\nRandom Forest Accuracy: {rf_accuracy:.4f}")
    
    print(f"\nTraining on group: {' - '.join(class_group)} with 5 features")
    print("Group size:", len(class_group))

    # Train KNN Classifier
    knn_clf = KNeighborsClassifier(n_neighbors=5, n_jobs=-1)
    # Select top features
    top_features = feature_importance.head(3)['feature'].tolist()
    X_train_selected = combined_train_clean[top_features]
    X_test_selected = combined_test_clean[top_features]
    X_val_selected = combined_val_clean[top_features]

    try:
        # Train and predict
        knn_clf.fit(X_train_selected, y_group)
        knn_predictions = knn_clf.predict(X_test_selected)
        knn_val_predictions = knn_clf.predict(X_val_selected)
        knn_accuracy = accuracy_score(test_labels_group, knn_predictions)
        
        print(f"KNN accuracy with 3 features: {knn_accuracy:.4f}")
        print("Top features used:", top_features)
        if knn_accuracy > best_accuracy:
            best_accuracy = knn_accuracy
            best_predictions = knn_predictions
            best_val_predictions = knn_val_predictions
        
    except Exception as e:
        print(f"Error with KNN: {str(e)}")
    
    if knn_predictions is None:
        print("Warning: No successful KNN models. Using RF predictions.")
        best_predictions = rf_predictions
        best_val_predictions = val_predictions
    
    # Train Bagging Classifier
    bag_clf = BaggingClassifier(
        n_estimators=400,
        max_samples=0.85,
        max_features=1.0,
        bootstrap=True,
        bootstrap_features=True,
        random_state=42,
        n_jobs=-1
    )
    # Select top features
    top_features = feature_importance.head(5)['feature'].tolist()
    X_train_selected = combined_train_clean[top_features]
    X_test_selected = combined_test_clean[top_features]
    X_val_selected = combined_val_clean[top_features]
    
    try:
        # Train and predict
        bag_clf.fit(X_train_selected, y_group)

        #print feature importance:
        if hasattr(bag_clf.estimators_[0], 'feature_importances_'):
            # Get feature importance from the first estimator
            bag_feature_importance = pd.DataFrame({
                'feature': X_train_selected.columns,
                'importance': bag_clf.estimators_[0].feature_importances_
            }).sort_values('importance', ascending=False)
            print(f"Feature importance: {bag_feature_importance.head(5)}")
        else:
            print("Feature importance not available for this estimator")

        bag_predictions = bag_clf.predict(X_test_selected)
        bag_val_predictions = bag_clf.predict(X_val_selected)
        bag_accuracy = accuracy_score(test_labels_group, bag_predictions)
        
        print(f"Bagging accuracy with 5 features: {bag_accuracy:.4f}")
        print("Top features used:", top_features)
        if bag_accuracy > best_accuracy:
            best_accuracy = bag_accuracy
            best_predictions = bag_predictions
            best_val_predictions = bag_val_predictions
            
    except Exception as e:
        print(f"Error with Bagging: {str(e)}")
    
    if bag_predictions is None:
        print("Warning: No successful bagging models. Using RF predictions.")
        best_predictions = rf_predictions
        best_val_predictions = val_predictions
    
    print(f"\nBest accuracy achieved: {best_accuracy:.4f}")
    
    return test_labels_group, best_predictions, best_accuracy, confusion_matrix(test_labels_group, best_predictions), best_val_predictions

def refine_and_retrain(X, y, test_df, test_labels, val_df, chronic_misclassifications, best_predictions, best_val_predictions,
                      social_media_features_train, social_media_features_test, social_media_features_val,
                      targeted_features_train, targeted_features_test, targeted_features_val,
                      enhanced_features_train, enhanced_features_test, enhanced_features_val):
    """
    Refine datasets by analyzing groups of related misclassified classes
    """
    print("\nRefining datasets...")

    # Convert best_predictions to pandas Series if it's a numpy array
    if isinstance(best_predictions, np.ndarray):
        best_predictions = pd.Series(best_predictions, index=test_labels.index)
    if isinstance(best_val_predictions, np.ndarray):
        best_val_predictions = pd.Series(best_val_predictions, index=val_df.index)
    
    # Get groups of related problematic classes
    class_groups = get_misclassified_groups(chronic_misclassifications)
    print(f"\nNumber of class groups identified: {len(class_groups)}")
    for i, group in enumerate(class_groups, 1):
        print(f"Group {i}: {' - '.join(group)}")
    
    # Store results for each group based on the best_predictions
    group_results = {}
    final_predictions = best_predictions.copy()
    final_val_predictions = best_val_predictions.copy()

    # Process each group
    for group in class_groups:
        print(f"\nProcessing group: {' - '.join(group)}")
        
        # Calculate initial accuracy for this group
        group_mask = best_predictions.isin(group)
        initial_group_labels = test_labels[group_mask]
        initial_group_predictions = best_predictions[group_mask]
        initial_group_accuracy = accuracy_score(initial_group_labels, initial_group_predictions)
        
        # Try to improve with specialized models
        group_labels, group_predictions, group_accuracy, group_cm, group_val_predictions = train_on_class_group(
            X, y, test_df, best_predictions, val_df, best_val_predictions, group,
            social_media_features_train, social_media_features_test, social_media_features_val,
            targeted_features_train, targeted_features_test, targeted_features_val,
            enhanced_features_train, enhanced_features_test, enhanced_features_val
        )

        is_improvement, metrics = evaluate_improvement(initial_group_labels, initial_group_predictions, group_predictions)
        
        print(f"New group accuracy: {group_accuracy:.4f}")
        
        # Only update predictions if we improved accuracy
        if is_improvement:
            print(f"True improvement achieved!")
            print(f"Accuracy change: {metrics['accuracy_change']:.4f}")
            print(f"Average recall change: {metrics['recall_change']:.4f}")
            
            # Update predictions
            test_mask = best_predictions.isin(group)
            val_mask = best_val_predictions.isin(group)
            
            final_predictions[test_mask] = group_predictions
            if sum(val_mask) > 0:
                final_val_predictions[val_mask] = group_val_predictions
                
            # Store results with improvement flag
            group_results[tuple(group)] = {
                'initial_accuracy': initial_group_accuracy,
                'final_accuracy': group_accuracy,
                'improvement': group_accuracy - initial_group_accuracy,
                'confusion_matrix': group_cm,
                'classification_report': classification_report(group_labels, group_predictions),
                'predictions_updated': True
            }
        else:
            print(f"No improvement for group {' - '.join(group)}. Keeping original predictions.")
            # Store results with no-improvement flag
            group_results[tuple(group)] = {
                'initial_accuracy': initial_group_accuracy,
                'final_accuracy': group_accuracy,
                'improvement': 0,
                'confusion_matrix': confusion_matrix(initial_group_labels, initial_group_predictions),
                'classification_report': classification_report(initial_group_labels, initial_group_predictions),
                'predictions_updated': False
            }
    
    # Print summary of group results
    print("\nGroup-wise Results Summary:")
    print("-" * 50)
    for group, results in group_results.items():
        print(f"\nGroup: {' - '.join(group)}")
        print(f"Initial accuracy: {results['initial_accuracy']:.4f}")
        print(f"Final accuracy: {results['final_accuracy']:.4f}")
        print(f"Improvement: {results['improvement']:.4f}")
        print(f"Predictions updated: {results['predictions_updated']}")
        if results['predictions_updated']:
            print("\nConfusion Matrix After Update:")
            print(results['confusion_matrix'])
            print("\nClassification Report After Update:")
            print(results['classification_report'])
    
    # Calculate and print final overall accuracy
    initial_accuracy = accuracy_score(test_labels, best_predictions)
    final_accuracy = accuracy_score(test_labels, final_predictions)
    print(f"\nOverall Results:")
    print(f"Initial accuracy: {initial_accuracy:.4f}")
    print(f"Final accuracy: {final_accuracy:.4f}")
    print(f"Overall improvement: {final_accuracy - initial_accuracy:.4f}")
    
    return final_predictions, final_val_predictions, group_results
```
