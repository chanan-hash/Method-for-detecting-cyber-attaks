```python
def analyze_data(df):
    """Analyze and visualize the dataset"""
    print("\nDataset Shape:", df.shape)
    print("\nMissing values:\n", df.isnull().sum())
    
    # Create correlation matrix for numeric columns
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    correlation_matrix = df[numeric_cols].corr()
    
    # Plot correlation matrix
    plt.figure(figsize=(15, 10))
    sns.heatmap(correlation_matrix, cmap='coolwarm', center=0, annot=False)
    plt.title('Feature Correlation Matrix')
    plt.tight_layout()
    plt.show()
    
    # Print class distribution if 'label' exists
    if 'label' in df.columns:
        print("\nClass Distribution:")
        print(df['label'].value_counts(normalize=True))
    
    return correlation_matrix
```

```python
def create_advanced_features(df):
    """Create advanced features from the raw data"""
    df = df.copy()
    
    # Advanced Time Features
    df['Timestamp'] = pd.to_datetime(df['Timestamp'], unit='s')
    df['hour'] = df['Timestamp'].dt.hour
    df['minute'] = df['Timestamp'].dt.minute
    df['day_of_week'] = df['Timestamp'].dt.dayofweek
    df['is_weekend'] = df['day_of_week'].isin([5, 6]).astype(int)
    df['is_business_hours'] = ((df['hour'] >= 9) & (df['hour'] <= 17)).astype(int)
    df['is_night'] = ((df['hour'] >= 23) | (df['hour'] <= 5)).astype(int)
    
    # IP-based features
    def ip_to_int(ip):
        try:
            parts = [int(part) for part in ip.split('.')]
            return parts[0] * 256**3 + parts[1] * 256**2 + parts[2] * 256 + parts[3]
        except:
            return 0
    
    df['Source_IP_num'] = df['Source_IP'].apply(ip_to_int)
    df['Destination_IP_num'] = df['Destination_IP'].apply(ip_to_int)
    
    # Basic Packet Features - check if columns exist first
    packet_features = []
    if 'fwd_packets_amount' in df.columns and 'bwd_packets_amount' in df.columns:
        df['total_packets'] = df['fwd_packets_amount'] + df['bwd_packets_amount']
        packet_features.append('total_packets')
        
    if 'fwd_packets_length' in df.columns and 'bwd_packets_length' in df.columns:
        df['total_bytes'] = df['fwd_packets_length'] + df['bwd_packets_length']
        packet_features.append('total_bytes')
        
        # Calculate packet ratios only if we have both length and amount
        if 'total_packets' in df.columns:
            df['avg_packet_size'] = df['total_bytes'] / (df['total_packets'] + 1)
            packet_features.append('avg_packet_size')
    
    # Protocol Features
    df['protocol_encoded'] = LabelEncoder().fit_transform(df['Protocol'])
    df = pd.get_dummies(df, columns=['Protocol'], prefix='protocol')
    
    # Flag Features - check if columns exist
    flag_cols = ['FIN_count', 'SYN_count', 'RST_count', 'PSH_count']
    if all(col in df.columns for col in flag_cols):
        df['total_flags'] = df['FIN_count'] + df['SYN_count'] + df['RST_count'] + df['PSH_count']
        if 'total_packets' in df.columns:
            df['flag_ratio'] = df['total_flags'] / (df['total_packets'] + 1)
    
    return df
```

```python
def select_features_advanced(X, y, correlation_threshold=0.95):
    """Advanced feature selection using multiple techniques"""
    # Remove constant and quasi-constant features
    selector = VarianceThreshold(threshold=0.01)
    X_var = pd.DataFrame(selector.fit_transform(X), columns=X.columns[selector.get_support()])
    
    # Remove highly correlated features using hierarchical clustering
    corr = spearmanr(X_var).correlation
    corr_linkage = hierarchy.ward(corr)
    cluster_ids = hierarchy.fcluster(corr_linkage, correlation_threshold, criterion='distance')
    cluster_id_to_feature_ids = defaultdict(list)
    for idx, cluster_id in enumerate(cluster_ids):
        cluster_id_to_feature_ids[cluster_id].append(idx)
    
    selected_features = [v[0] for v in cluster_id_to_feature_ids.values()]
    X_uncorr = X_var.iloc[:, selected_features]
    
    # Use RFECV for final feature selection
    rf = RandomForestClassifier(n_estimators=100, random_state=42)
    selector = RFECV(rf, step=1, cv=5, scoring='accuracy', n_jobs=-1)
    selector.fit(X_uncorr, y)
    
    return X_uncorr.iloc[:, selector.support_]
```

```python

def create_advanced_ensemble():
    """Create an advanced ensemble with multiple base models"""
    # Base models
    base_rf = RandomForestClassifier(n_estimators=200, random_state=42)
    base_xgb = XGBClassifier(n_estimators=200, random_state=42)
    base_lgbm = LGBMClassifier(n_estimators=200, random_state=42)
    
    # Create bagging classifiers
    bagged_rf = BaggingClassifier(
        estimator=DecisionTreeClassifier(),
        n_estimators=100,
        random_state=42
    )
    
    # Define base models for stacking
    estimators = [
        ('rf', base_rf),
        ('xgb', base_xgb),
        ('lgbm', base_lgbm),
        ('bagged_rf', bagged_rf)
    ]
    
    # Create stacking classifier
    stacking = StackingClassifier(
        estimators=estimators,
        final_estimator=GradientBoostingClassifier(n_estimators=100, random_state=42),
        cv=5,
        n_jobs=-1
    )
    
    return stacking
```

```python
def train_advanced_model(X_train, X_test, y_train, y_test):
    """Train the advanced ensemble model"""
    # Create and train stacking model
    model = create_advanced_ensemble()
    
    # Define parameter grid
    param_grid = {
        'rf__max_depth': [20, 30],
        'rf__min_samples_split': [2, 5],
        'xgb__max_depth': [6, 8],
        'xgb__learning_rate': [0.01, 0.1],
        'lgbm__num_leaves': [31, 63]
    }
    
    # Perform grid search with stratification
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    grid_search = GridSearchCV(model, param_grid, cv=cv, scoring='accuracy', n_jobs=-1)
    
    print("Training ensemble model...")
    grid_search.fit(X_train, y_train)
    
    # Get best model and make predictions
    best_model = grid_search.best_estimator_
    y_pred = best_model.predict(X_test)
    
    # Print evaluation metrics
    print("\nBest parameters:", grid_search.best_params_)
    print("\nModel Performance:")
    print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))
    
    return best_model
```

```python
def main():
    """Main execution pipeline"""
    # Load data
    print("Loading data...")
    train_df = pd.read_csv('../APP-1/train.csv')
    test_df = pd.read_csv('../APP-1/test.csv')
    
    # Analyze data
    print("\nAnalyzing training data...")
    analyze_data(train_df)
    
    # Create advanced features
    print("\nCreating advanced features...")
    train_df_processed = create_advanced_features(train_df)
    test_df_processed = create_advanced_features(test_df)
    
    # Prepare target variable
    y_train = train_df_processed['label']
    y_test = test_df_processed['label']
    
    # Encode labels
    label_encoder = LabelEncoder()
    y_train_encoded = label_encoder.fit_transform(y_train)
    y_test_encoded = label_encoder.transform(y_test)
    
    # Prepare features
    drop_cols = ['label', 'Source_IP', 'Destination_IP', 'Timestamp']
    X_train = train_df_processed.drop(columns=drop_cols)
    X_test = test_df_processed.drop(columns=drop_cols)
    
    # Advanced feature selection
    print("\nPerforming feature selection...")
    X_train_selected = select_features_advanced(X_train, y_train_encoded)
    X_test_selected = X_test[X_train_selected.columns]
    
    # Scale features
    print("\nScaling features...")
    scaler = RobustScaler()
    X_train_scaled = scaler.fit_transform(X_train_selected)
    X_test_scaled = scaler.transform(X_test_selected)
    
    # Train model
    best_model = train_advanced_model(X_train_scaled, X_test_scaled, 
                                    y_train_encoded, y_test_encoded)
    
    # Save selected features
    selected_features = X_train_selected.columns.tolist()
    
    return best_model, selected_features
```

```python
if __name__ == "__main__":
    best_model, selected_features = main()
```
